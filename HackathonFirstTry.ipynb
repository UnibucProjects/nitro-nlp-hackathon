```python
import json
import pandas as pd
```

```python
path = './train.json'
```

```python
continuturi = pd.read_json(path)
```

```python
continuturi
```

```python
continuturi.describe()
```

```python
continuturi.head()
```

```python
n = len(continuturi)
sentences = []
for i in range(n):
    sentence = ""
    for j, token in enumerate(continuturi["tokens"][i]):
        sentence += token
        if continuturi["space_after"][i][j]:
            sentence += " "
    sentences.append(sentence)

continuturi["sentences"] = sentences

continuturi.head()
```

```python
continuturi['sentences'][0]
```

```python
continuturi['sentences'][1]
```

```python
! pip install simpletransformers
```

```python
sentences_id = []
words = []
tags = []

for i in range(n):
  for j, token in enumerate(continuturi["tokens"][i]):
    sentences_id.append(i)
    words.append(token)
    tags.append(continuturi["ner_tags"][i][j])

print(sentences_id[:10])
print(words[:10])
print(tags[:10])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
train_data = pd.DataFrame({"sentence_id":sentences_id, "words": words, "labels":tags})

train_data
```

```python
from simpletransformers.ner import NERModel, NERArgs
```

```python
labels = train_data['labels'].unique().tolist()
labels
```

```python
args = NERArgs()
args.num_train_epochs = 1
args.learning_rate = 1e-4
args.overwrite_output_dir = True
args.train_batch_size = 32
args.eval_batch_size = 32
```

```python
model = NERModel('bert', 'bert-base-cased', labels=labels, args=args)
```

```python
from sklearn.metrics import accuracy_score
```

```python
model.train_model(train_data, acc=accuracy_score)
```

```python
import pickle
pickle.dump(model, open('model.pkl', 'wb'))
```

```python
prediction, model_output = model.predict(['În România, ca de altfel în întreaga Europă, stand up-ul este forma de divertisment cea mai îndrăgită de tineri în acest moment, comediantii reușind să umple săli care erau destinate mai degrabă artiștilor din zona muzicii pop rock, iar Badea, Bordea și Micutzu sunt printre cei care au contribuit direct la succesul acestei forme de divertisment în țara noastră.', 'Se estimează că în prezent acestea sunt utilizate de mai mult de 50 de milioane de persoane fizice și juridice din lumea întreagă.'])
```

```python
test_path = './test.json'
test_content = pd.read_json(test_path)
```

```python
test_content
```

```python
n_test = len(test_content)
sentences = []
for i in range(n_test):
    sentence = ""
    for j, token in enumerate(test_content["tokens"][i]):
        if token == '\n' or token == '\n ':
          sentence += 'endl'
        elif token == ' ':
          sentence += 'space'
        else:
          sentence += token
        sentence += " "
    sentences.append(sentence[:-1])

test_content["sentences"] = sentences

test_content.head()
```

```python
tag_to_id = json.load(open('tag_to_id.json'))
```

```python
tag_to_id
```

```python
all_predictions = []

predictions, model_output = model.predict(list(test_content['sentences']))
```

```python
print(len(predictions))
```

```python
gresite = 0
for i, prediction in enumerate(predictions):
  if len(prediction) != len(test_content['tokens'][i]):
    print(i, len(prediction), len(test_content['tokens'][i]))

print(gresite)
```

```python
print(test_content['sentences'][1438])
print(test_content['tokens'][1438])
print(predictions[1438])
```

```python
for i,prediction in enumerate(predictions):
  sentence_prediction = []
  if len(prediction) == len(test_content['tokens'][i]):
    for j in range(len(prediction)):
      sentence_prediction.append(tag_to_id[list(prediction[j].values())[0]])
  else:
    for token in test_content['tokens'][i]:
      for p in prediction:
        if token == list(p.keys())[0]:
          sentence_prediction.append(tag_to_id[list(p.values())[0]])
          break
      else:
        sentence_prediction.append(tag_to_id['O'])
  all_predictions.append(sentence_prediction)

```

```python
gresite = 0
for i in range(len(all_predictions)):
  if len(all_predictions[i]) != len(test_content['tokens'][i]):
    gresite += 1

print(gresite)
```

```python
all_token_pred = []
for sentence in all_predictions:
  for token_pred in sentence:
    all_token_pred.append(token_pred)

total = 0
for tokens in test_content['tokens']:
  total += len(tokens)

print(len(all_token_pred))
print(total)
```

```python
import csv

header = ['Id', 'ner_label']

with open('submission.csv', 'w') as f:
  writer = csv.writer(f)
  writer.writerow(header)
  for i, pred_label in enumerate(all_token_pred):
    writer.writerow([i, pred_label])
  
```
